```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
spark = SparkSession.builder \
.master("local") \
.appName("loop_example_df") \
.getOrCreate()
```

```python
df1 = spark.read \
.format("csv") \
.option("header", True) \
.option("inferSchema", True) \
.option("path","D:\BigData\dataset\simple-zipcodes.csv") \
.load()
```

```python
df1.write \
.format("csv") \
.partitionBy("state") \
.mode("Overwrite") \
.option("path","C:\\Users\\ASUS\\pyspark\\Untitled Folder\\output") \
.save() 
```
